```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
marks = [8,10,15,20]
new_min = 0
new_max = 1
mean = np.mean(marks)
std_dev = np.std(marks)

print(f"mean marks {mean}")
print(f"standard dev marks {std_dev}")
```

```python
def z_score_normalisation(marks,mean,std_dev):
  for mark in marks:
    print(f"Z score normalisation for {mark} is {(mark-mean)/std_dev}")


def min_max_normalisation(marks,new_max,new_min):
  min_a = min(marks)
  max_a = max(marks)
  for mark in marks:
    minmax = ((mark - min_a)/(max_a - min_a)) * (new_max - new_min) + new_min
    print(f"Minmax normalisation for {mark} is {minmax}")
```

```python

min_max_normalisation(marks,new_max,new_min)
print("\n")
z_score_normalisation(marks,mean,std_dev)
```

Iris Dataset

```python
iris_data = load_iris()
sepal_len = []
sepal_width = []
petal_len = []
petal_width = []

#seperate the features of iris
for data in iris_data.data:
  sepal_len.append(data[0])
  sepal_width.append(data[1])
  petal_len.append(data[2])
  petal_width.append(data[3])

#for sepal length feature of iris
mean_sepal_len = np.mean(sepal_len)
std_dev_sepal_len = np.std(sepal_len)

z_score_normalisation(sepal_len,mean_sepal_len,std_dev_sepal_len)
min_max_normalisation(sepal_len,new_max,new_min)
```
