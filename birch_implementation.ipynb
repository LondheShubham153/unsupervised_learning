Cluster the Above Data Using BIRCH Algorithm, considering T<1.5>, and Max Branch = 2 

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
```

```python
X, clusters = make_blobs(n_samples=1000, centers=12, cluster_std=0.50, random_state=0)
plt.scatter(X[:,0], X[:,1], alpha=0.7, edgecolors='b')
```

```python
model = Birch(branching_factor=50, n_clusters=None, threshold=1.5)
model.fit(X)
```

```python
predict = model.predict(X)
predict
```

```python
plt.scatter(X[:,0], X[:,1], c=predict, cmap='rainbow', alpha=0.7, edgecolors='b')
```

```python
data = np.array([[2,6],[4,5],[4,7],[6,2],[7,2],[8,4],[7,9]])
data
```

```python
plt.scatter(data[:,0], data[:,1], alpha=0.7, edgecolors='b')
```

```python
model = Birch(branching_factor=2, n_clusters=None, threshold=1.5)
model.fit(data)
```

```python
pred = model.predict(data)
pred
```

```python
plt.scatter(data[:,0], data[:,1], c=pred, cmap='rainbow', alpha=0.7, edgecolors='b')
```
